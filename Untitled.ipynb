```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain_community.vectorstores import Chroma

long_conversation = """
User asked about their project deadline: The science project is due on March 20th.
User mentioned their teacher's name: The teacher's name is Mr. Johnson.
User said their partner's name: Their project partner is named Sarah.
User mentioned the project topic: The project is about renewable energy.
User shared their school name: They go to Foothill High School.
User mentioned their grade: They are in 10th grade.
User said their favorite part: Their favorite part of the project is the solar panel section.
User mentioned a concern: They are worried about the presentation on March 21st.
User shared a fun fact: Sarah is really good at making slides.
User mentioned their goal: They want to get an A on the project.
"""

splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20)
chunks = splitter.split_text(long_conversation)
print(f"Created {len(chunks)} chunks:")
for i, chunk in enumerate(chunks):
    print(f"  Chunk {i+1}: {chunk[:60]}...")

#The problem is that this takes very long time, around 1-2 miunutes
```

```python
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = Chroma.from_texts(chunks, embeddings)
print("Vector database created!")

query = "What is the project about?"
results = vectorstore.similarity_search(query, k=2)

print(f"\nTop results for: '{query}'")
for i, result in enumerate(results):
    print(f"  Result {i+1}: {result.page_content}")
```

```python
def ask_with_rag(question):
    results = vectorstore.similarity_search(question, k=2)
    context = "\n".join([r.page_content for r in results])
    
    prompt = f"""You are a helpful assistant. Only use the context you are provided with to answer the questions. If you don't know the answer, reply with I don't know."

Context:
{context}

Question: {question}
Answer:"""
    
    response = ollama.chat(
        model="llama3.2",
        messages=[{"role": "user", "content": prompt}]
    )
    
    print(f"Question: {question}")
    print(f"Retrieved context: {context}")
    print(f"Answer: {response['message']['content']}")
    print("-" * 50)

ask_with_rag("What is the teacher's name?")
ask_with_rag("When is the presentation?")
ask_with_rag("Who is the project partner?")
```

```python
import json

# Load the dataset
with open("train-v1.1.json") as f:
    squad = json.load(f)

passages = []
qa_pairs = []

for article in squad["data"][:5]: 
    for paragraph in article["paragraphs"]:
        passages.append(paragraph["context"])
        for qa in paragraph["qas"][:2]: 
            qa_pairs.append({
                "question": qa["question"],
                "answer": qa["answers"][0]["text"] 
            })

print(f"Loaded {len(passages)} passages")
print(f"Loaded {len(qa_pairs)} Q&A pairs")
print("\nExample Q&A:")
print(f"Q: {qa_pairs[0]['question']}")
print(f"A: {qa_pairs[0]['answer']}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain_community.vectorstores import Chroma

# Split passages into chunks and store in ChromaDB
splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

all_chunks = []
for passage in passages:
    chunks = splitter.split_text(passage)
    all_chunks.extend(chunks)

vectorstore = Chroma.from_texts(all_chunks, embeddings)
print(f"Stored {len(all_chunks)} chunks in ChromaDB")
```

```python
def ask_baseline(question):
    response = ollama.chat(
        model="llama3.2",
        messages=[{"role": "user", "content": f"Answer this question in a few words only: {question}"}]
    )
    return response['message']['content'].strip()

def ask_with_rag_scored(question):
    results = vectorstore.similarity_search(question, k=3)
    context = "\n".join([r.page_content for r in results])
    prompt = f"""Answer using only the context below. Be brief and concise, try to answer in short words/sentences.
If the answer isn't in the context, say "I don't know."

Context:
{context}

Question: {question}
Answer:"""
    response = ollama.chat(
        model="llama3.2",
        messages=[{"role": "user", "content": prompt}]
    )
    return response['message']['content'].strip()

test_pairs = qa_pairs[:10]
baseline_scores = []
rag_scores = []

print("Running tests... this will take a minute\n")
print(f"{'Question':<50} {'Correct Answer':<30} {'Baseline':<30} {'RAG':<30}")
print("-" * 140)

for qa in test_pairs:
    q = qa["question"]
    correct = qa["answer"].lower()
    
    baseline_ans = ask_baseline(q)
    rag_ans = ask_with_rag_scored(q)
    
    baseline_correct = 1 if correct in baseline_ans.lower() else 0
    rag_correct = 1 if correct in rag_ans.lower() else 0
    
    baseline_scores.append(baseline_correct)
    rag_scores.append(rag_correct)
    
    print(f"{q[:48]:<50} {correct[:28]:<30} {baseline_ans[:28]:<30} {rag_ans[:28]:<30}")

print("\n" + "=" * 140)
print(f"BASELINE SCORE: {sum(baseline_scores)}/10")
print(f"RAG SCORE:      {sum(rag_scores)}/10")
print(f"IMPROVEMENT:    +{sum(rag_scores) - sum(baseline_scores)} questions answered correctly")
```
